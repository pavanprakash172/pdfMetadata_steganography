```python
%%writefile LSBsteg.py
import PyPDF2 as pd
import argparse

#Decryption
def decryption(file_path):
    reader=pd.PdfReader(file_path)
    metadata=reader.metadata
        
    for key in metadata:
        print(f'{key}:{metadata[key]}')
            

#Encryption
def encryption(file_path,metadata_keyName,message,outputfile):
    with open(file_path,'rb') as file:
        reader=pd.PdfReader(file_path)
        writer=pd.PdfWriter()
        writer.append_pages_from_reader(reader)
    
        metadata=reader.metadata
        metadata.update({metadata_keyName:message})
        writer.add_metadata(metadata)
    
        with open(outputfile ,'wb') as file:
            writer.write(file)
            print("message embedded succesfully.")

def main():
    parser=argparse.ArgumentParser(description="this program will emded or decode the hidden message")
    args,unknown=parser.parse_known_args()
    parser.add_argument('-d', action="store_true",help='decodes the hidden data.')
    parser.add_argument('-e', action="store_true",help='embed a hidden data.')
    parser.add_argument('-f', required=True,help='file location.')
    parser.add_argument('-o', required=False,help='name of the output file.')
    parser.add_argument('-mn', required=False,help='metadata keyName.')
    parser.add_argument('-m', required=False,help='message or data that needs to embed into the file.')

    args=parser.parse_args()

    if args.d:
        decryption(args.f)

    if args.e:
        if args.mn and args.m and args.o:
            encryption(args.f,args.mn,args.m,args.o)
        else:
            parser.error("This process needs -mn, -m, -o flags.")
            

if __name__=="__main__":
    main()    
```
